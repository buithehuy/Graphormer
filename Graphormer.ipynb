```
import os
import site

# TÃ¬m táº¥t cáº£ cÃ¡c thÆ° má»¥c site-packages trong mÃ´i trÆ°á»ng
possible_paths = [
    '/usr/local/envs/graphormer/lib/python3.9/site-packages',
    '/usr/local/lib/python3.9/site-packages'
]

# Ná»™i dung báº£n vÃ¡
patch_code = """
import os
import requests

class GoogleDriveDownloader:
    @staticmethod
    def download_file_from_google_drive(file_id, dest_path, overwrite=False, showsize=False):
        destination_directory = os.path.dirname(dest_path)
        if not os.path.exists(destination_directory):
            os.makedirs(destination_directory)
        if not os.path.exists(dest_path) or overwrite:
            URL = "https://docs.google.com/uc?export=download"
            session = requests.Session()
            response = session.get(URL, params={'id': file_id}, stream=True)
            token = None
            for key, value in response.cookies.items():
                if key.startswith('download_warning'):
                    token = value
                    break
            if token:
                params = {'id': file_id, 'confirm': token}
                response = session.get(URL, params=params, stream=True)
            CHUNK_SIZE = 32768
            with open(dest_path, "wb") as f:
                for chunk in response.iter_content(CHUNK_SIZE):
                    if chunk: f.write(chunk)
"""

found = False
for base_path in possible_paths:
    # Thá»­ cáº£ hai tÃªn thÆ° má»¥c phá»• biáº¿n
    for folder_name in ['google_drive_downloader', 'googledrivedownloader']:
        target_dir = os.path.join(base_path, folder_name)
        if os.path.exists(target_dir):
            init_file = os.path.join(target_dir, '__init__.py')
            with open(init_file, 'w') as f:
                f.write(patch_code)
            print(f"--- ÄÃ£ vÃ¡ thÃ nh cÃ´ng táº¡i: {init_file}")
            # Táº¡o thÃªm symlink náº¿u cáº§n Ä‘á»ƒ cháº¯c cháº¯n PyG tÃ¬m tháº¥y
            link_path = os.path.join(base_path, 'google_drive_downloader')
            if not os.path.exists(link_path):
                os.symlink(target_dir, link_path)
            found = True
            break
    if found: break

if not found:
    print("âŒ KhÃ´ng tÃ¬m tháº¥y thÆ° má»¥c cÃ i Ä‘áº·t. Cháº¡y lá»‡nh '!conda run -n graphormer pip install googledrivedownloader'")

# Kiá»ƒm tra láº¡i ngay láº­p tá»©c
!conda run -n graphormer python -c "from google_drive_downloader import GoogleDriveDownloader; print('---'); print('OK!')"
```
